```python
# 특정문자(?)로 인해 문자형으로 지정된 경우
# => 문자를 결측자료로 변경 
# => 결측자료 제거
```

```python
import numpy as np
df['horsepower'].replace('?', np.nan)
```

```python
# .replace(대상값, 변경할 값)
# .replace({대상값1:변경할 값1, 대상값2:변경할 값2, ...})

df['horsepower'] = df['horsepower'].replace('?', np.nan)
df.count()
```

```python
# 자료 유형 변경
# 자료 유형 : 정수형(int), 실수형(float), 문자형(str)
df['horsepower'] = df['horsepower'].astype('float')
```
