```python
# 객체 = LogisticRegression(옵션들).fit(독립, 종속)
# 객체.coef_ : 회귀계수
# 객체.predict(독립) : 예측값(분류된)
# 객체.predict_proba(독립) : 예측값(확률)
# !!옵션들
# penalty = 'L2', 'L1', 'none'(통계패키지와 같은 결과) => penalty를 적용한 Logistic Regression이 default
# tol = 수렴 판단 기준값
# solver = 수치해석학 방법(Newton-Raphson etc. .....)
```

```python
import seaborn as sns
titanic = sns.load_dataset('titanic')
titanic.info()
```

```python
# 종속 : survived
# 독립 : fare(연속), pclass(범주)
data = pd.get_dummies(titanic, columns=['class'])
```

```python
data.info()
```

```python
y = data['survived']
X = data[['fare', 'class_First', 'class_Second', 'class_Third']]
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0, stratify=y)
```

```python
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression(penalty=None).fit(X_train, y_train)
```

```python
LR.coef_
```

```python
y_hat1 = LR.predict(X_test)
```

```python
LR.predict_proba(X_test) # 예측된 확률값 (0.5 기준으로 분류)
```

```python
# odd ratio
import numpy as np
np.exp(LR.coef_)
```

```python
y_prob = LR.predict_proba(X_test)[:,1] # y가 1이 될 확률만 추출 P(y=1)
```

```python
# np.where(조건, 참일때, 거짓일때)
import numpy as np
y_hat2 = np.where(y_prob > 0.4, 1, 0)
```

```python
# y # 실제값
# y_hat1 # 예측값
# y_hat2 # 예측값
```

```python
from sklearn.metrics import confusion_matrix
# confusion_matrix(실제값, 예측값)
```

```python
confusion_matrix(y_test, y_hat1)
```

```python
confusion_matrix(y_test, y_hat2)
```

```python
from sklearn.metrics import classification_report
# classification_report(실제값, 예측값)
```

```python
print(classification_report(y_test, y_hat1))
```

```python
print(classification_report(y_test, y_hat2))
```

```python
# LabelEncoder
data = mpg.loc[mpg['origin'] != 'europe']
from sklearn.preprocessing import LabelEncoder
classle = LabelEncoder()
data['origin'] = classle.fit_transform(data['origin']) # 넘버링 기준 --> a(0) => z(1)
```

```python
y = data['origin']
X = data[['mpg', 'weight']]
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0, stratify=y)
```

```python
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression(penalty=None).fit(X_train, y_train)
```

```python
y_test_hat = LR.predict(X_test) # 0.5의 확률 기준으로 분류
```

```python
from sklearn.metrics import confusion_matrix, classification_report
confusion_matrix(y_test, y_test_hat)
```

```python
print(classification_report(y_test, y_test_hat))
```

```python
# ROC curve
from sklearn.metrics import roc_curve, auc
```

```python
y_train_prob = LR.predict_proba(X_train)[:,1]
# fpr, tpr, _ = roc_curve(실제값, y=1 일 확률)
fpr, tpr, _ = roc_curve(y_train, y_train_prob)
roc_auc = auc(fpr, tpr)
```

```python
import matplotlib.pyplot as plt
# plt.plot(x축값, y축값, label=범례 내용)
plt.plot(fpr, tpr, 'b', label = 'AUC=%.2f' %roc_auc)
plt.plot([0,1],[0,1],'--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC curve')
plt.legend(loc='best')
```
